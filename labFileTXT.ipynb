# работа с текстовыми файлами
r открывает для чтения если нет ошибка
w только для записи файл будет очищен создаётся если нет
x то же самое но ошибка если нет
a запись в конец если нет то создаётся

_t текстовый фал по умолч
_b бинарные
_+ r+w

```python
import os
#временный каталог
TEMP_DIR='temp'
if not os.path.exists(TEMP_DIR):
    os.mkdir(TEMP_DIR)
```

```python
text="HI THIS IS TEXT AND YOU ARE READING IT\n"
```

# ЗАПИСЬ В ФАЙЛ
\n\n\n\n\n\n\n\\n\n

```python
# через try-finally
f=open(os.path.join(TEMP_DIR, 'temp_TXT.txt'), 'w')
try:
    f.write(text)
    f.write(text)
    f.write(text)
    f.write(text)
    f.write(text)
    f.write(text)
    f.write(text)
finally:
    f.close()
```

```python
# запись через with as
with open (os.path.join(TEMP_DIR, 'temp_TXT.txt'),'w') as f:
    for i in range(100):
        f.write(text)
```

```python
with open (os.path.join(TEMP_DIR, 'temp2_TXT.txt'),'w') as f:
    f.write('\n'.join(f'линия {i}' for i in range(100)))
```

```python
with open (os.path.join(TEMP_DIR, 'temp2_TXT.txt'),'r') as f:
    print(f.read())
```

```python
with open (os.path.join(TEMP_DIR, 'temp2_TXT.txt'),'r') as f:
    for line in f:
        print(f"{line=}")
```

```python
with open (os.path.join(TEMP_DIR, 'temp2_TXT.txt'),'r') as f:
    print(f.readlines())
```

```python
with open (os.path.join(TEMP_DIR, 'temp2_TXT.txt'),'r') as f:
    for n,line in enumerate(f):
        print(f"{repr(line)}  {n}")
```
